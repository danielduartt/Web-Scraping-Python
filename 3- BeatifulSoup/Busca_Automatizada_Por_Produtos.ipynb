```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

##**Objetivo:**
<p> Obter dados de produtos do Mercado Livre a partir de uma busca realizada pelo o usuário </p>

#Primeiro procurando apenas um produto


```python
url_base = 'https://lista.mercadolivre.com.br/'
produto = input("Qual o produto você deseja procurar: ").strip().upper()
response = requests.get(url_base + produto)
if response.status_code == 200:
  print("Produto encontrado")
else:
  print("Algo deu errado, Tente novamente")
```

```python
pagina_produto = BeautifulSoup(response.text, "html.parser")

produtos = pagina_produto.find("div", attrs = {"class": "ui-search-result__wrapper shops__result-wrapper"})
```

```python
titulo_produto = produtos.find('h2', attrs = {"class": "ui-search-item__title"})
preco_produto = produtos.find("div", attrs = {"class": "ui-search-price__second-line shops__price-second-line"})
valor = preco_produto.find('span', attrs = {'class': "andes-money-amount__fraction"})
link_produto = produtos.find('a', attrs = {"class": "ui-search-link" })
link_produto = link_produto['href']
```

```python
print(titulo_produto.text)
print(f"Valor: {valor.text}")
print(f"Link:\n{link_produto}")
```

#Procurando por todos os Produtos

```python
produtos_all = pagina_produto.findAll("div", attrs = {"class": "ui-search-result__wrapper shops__result-wrapper"})
lista_produtos = list()
for produtos in produtos_all:
  titulo_produto = produtos.find('h2', attrs = {"class": "ui-search-item__title"})
  preco_produto = produtos.find("div", attrs = {"class": "ui-search-price__second-line shops__price-second-line"})
  valor = preco_produto.find('span', attrs = {'class': "andes-money-amount__fraction"}).text
  link_produto = produtos.find('a', attrs = {"class": "ui-search-link" })
  link_produto = link_produto['href']
  lista_produtos.append([titulo_produto, valor, link_produto])

produtos_df = pd.DataFrame(lista_produtos, columns = ['Título', "Preço", "Link"])
```
